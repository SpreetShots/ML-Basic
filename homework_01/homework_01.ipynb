## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
while True:
    print('\033[96m' + "Отправьте q для выхода из программы" + '\033[0m')
    text = input("Введите 5 значное число: ")
    if text.__contains__("q"):
        print('\033[93m' + "Завершение выполнения кода" + '\033[0m')
        break
    else:
        if len(text) == 5 and text.isdigit():
            swapped_str = text[0] + text[3] + text[2] + text[1] + text[4]
            print('\033[92m' + f"Результат: {text} -> {swapped_str}" + '\033[0m')
        else:
            print('\033[91m' + "Не введо верное 5 значное число" + '\033[0m')
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
while True:
    print('\033[96m' + "Отправьте q для выхода из программы" + '\033[0m')
    
    num_day = input("Введите дней до отпуска: ")
    
    if num_day.__contains__("q"):
        break
    
    if num_day.isdigit():
        int_num_day = int(num_day)
        sunday = int_num_day // 7
        saturday = (int_num_day + 1) // 7
        count_of_days_off = sunday + saturday
        print('\033[92m' + f"{num_day} -> {count_of_days_off}" + '\033[0m')
    else:
        print('\033[91m' + "Введите корректное числовое значение" + '\033[0m')
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
width = input("Введите ширину плитки: ")
height = input("Введите высоту плитки: ")
length_plitki = input("Введите размер куска плитки: ")

if width.isdigit() and height.isdigit() and length_plitki.isdigit():
    
    int_width = int(width)
    int_height = int(height)
    int_length_plitki = int(length_plitki)
    
    check_width = int_length_plitki % int_width
    check_height = int_length_plitki % int_height
    check_count_width = int_length_plitki // int_width
    check_count_height = int_length_plitki // int_height
    
    if check_width == 0 and check_height == 0:
        if check_count_width <= int_height:
            print('\033[92m' + f"{width}, {height}, {length_plitki} -> True" + '\033[0m')
        elif check_count_height <= int_width:
            print('\033[92m' + f"{width}, {height}, {length_plitki} -> True" + '\033[0m')
        else:
            print('\033[93m' + f"{width}, {height}, {length_plitki} -> False" + '\033[0m')
    elif check_width == 0:
        if check_count_width <= int_height:
            print('\033[92m' + f"{width}, {height}, {length_plitki} -> True" + '\033[0m')
        else:
            print('\033[93m' + f"{width}, {height}, {length_plitki} -> False" + '\033[0m')
    elif check_height == 0:
        if check_count_height <= int_width:
            print('\033[92m' + f"{width}, {height}, {length_plitki} -> True" + '\033[0m')
        else:
            print('\033[93m' + f"{width}, {height}, {length_plitki} -> False" + '\033[0m')
    else:
        print('\033[93m' + f"{width}, {height}, {length_plitki} -> False" + '\033[0m')
else:
    print('\033[91m' + "Введите корректные числа" + '\033[0m')
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
while True:
    
    print('\033[96m' + "Отправьте q для выхода из программы" + '\033[0m')    
    
    numbers_arabic = input("Введите число: ")
    numbers_roman = ""
    
    
    if numbers_arabic.__contains__("q"):
        break
    
    if numbers_arabic.isdigit() and int(numbers_arabic) < 4000:
        int_numbers_arabic = int(numbers_arabic)
        while int_numbers_arabic >= 1000:
            numbers_roman += 'M'
            int_numbers_arabic -= 1000
        if int_numbers_arabic >= 900:
            numbers_roman += 'CM'
            int_numbers_arabic -= 900
        if int_numbers_arabic >= 500:
            numbers_roman += 'D'
            int_numbers_arabic -= 500
        if int_numbers_arabic >= 400:
            numbers_roman += 'CD'
            int_numbers_arabic -= 400
        while int_numbers_arabic >= 100:
            numbers_roman += 'C'
            int_numbers_arabic -= 100
        if int_numbers_arabic >= 90:
            numbers_roman += 'XC'
            int_numbers_arabic -= 90
        if int_numbers_arabic >= 50:
            numbers_roman += 'L'
            int_numbers_arabic -= 50
        if int_numbers_arabic >= 40:
            numbers_roman += 'XL'
            int_numbers_arabic -= 40
        while int_numbers_arabic >= 10:
            numbers_roman += 'X'
            int_numbers_arabic -= 10
        if int_numbers_arabic >= 9:
            numbers_roman += 'IX'
            int_numbers_arabic -= 9
        if int_numbers_arabic >= 5:
            numbers_roman += 'V'
            int_numbers_arabic -= 5
        if int_numbers_arabic >= 4:
            numbers_roman += 'IV'
            int_numbers_arabic -= 4
        while int_numbers_arabic >= 1:
            numbers_roman += 'I'
            int_numbers_arabic -= 1
            
        print('\033[92m' + f"{numbers_arabic} -> {numbers_roman}" + '\033[0m')
    else:
        print('\033[91m' + "Введите корректное число в диапазоне от 0 до 3999" + '\033[0m')
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
while True:
    
    print('\033[96m' + "Отправьте q для выхода из программы" + '\033[0m')
    
    numbers_float = input("Введите число: ")
    
    if numbers_float.__contains__("q"):
        break
    
    if numbers_float != '':
        is_float = True
        has_point = False
        is_first_char = True

        for value in numbers_float:
            if value == '.':
                if has_point:
                    is_float = False
                    break
                has_point = True
            elif value < '0' or value > '9':
                if is_first_char and value == '-':
                    is_float = False
                    break
                is_float = False
                break
            is_first_char = False

        if numbers_float[0] == '0' and len(numbers_float) > 1 and not has_point:
            is_float = False
        if is_float and (numbers_float[0] == '-' or (has_point and numbers_float.find('.') == 0)):
            is_float = False

        print(f"{numbers_float} -> {is_float}")
    else:
        print('\033[91m' + "Введите значение" + '\033[0m')
```
